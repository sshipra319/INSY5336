Problem 1: 

1) In main() function ask the user the name of the file, opening the file and handle the error while opening the file. 
2) Initialize word count and character count to 0.
3) Traverse through each line of the file, split the words and count the words. Print the number of words in file.
4) Traverse through each word and count the number of characters. Print the number of characters in the file.
5) Average length of a word is calculated by dividing the character count to the word count.

```python
def main():
    filename = input("Enter the name of the file: ")  # Asking for the file name
    while(True):
        try:              # Checking the file availability - error catching
            fhand = open(filename, 'r')
            break
        except:
            print("File cannot be opened: ", filename)
            filename = input("Please enter a valid file name: ")
    
    lines = fhand.readlines()          #reading lines of file
    fhand.close()
    word_count = 0              # count the words
    char_count = 0              # count the characters
    
    for line in lines:
        words = line.strip().split()            # for each line split words
        word_count = word_count + len(words)     # count words
        for word in words:                        
            char_count = char_count + len(word)   #count characters in each word
    
    average = char_count/word_count
    print("Number of lines: ", len(lines))
    print("Number of words: ", word_count)
    print("Number of characters: ", char_count)      # print the values
    print("Average length of word: ", average)         
    

if __name__ == "__main__":
    main()
    
    
```

Problem 2:

1) In main() function ask the user the name of the file, opening the file and handle the error while opening the file.
2) The user is asked for the input using the choice variable. If user inputs choice = 1, ask the name of the reviewers, check the reviewers names in the file, if not found throw an error. If reviewer is found in the file, euclidean distance is calculated and user is again asked for choices.
3) If user inputs choice = 2, ask user the name of the reviewer, check the reviewer name in the file, if not found throw an error. If found in the file, calculate the euclidean distance with each entry in the file.
4) If user inputs choice = 3, it prints Goodbye!

```python
import csv
from math import sqrt


def main():
    movie_review = dict()
    filename = input("Enter the name of the file: ")  # Asking for the file name
    while True:
        try:  # Checking the file availability - error catching
            with open(filename, 'r') as revfile:
                readfile = csv.reader(revfile)       # redaing the csv file
                for row in readfile:
                    if len(row) == 0:
                        break
                    else:
                        rlist = row[1:]  # Create a dictionary of entries
                        rdict = {rlist[i]: rlist[i+1] for i in range(0,len(rlist),2)}
                        name = (row[0].strip()).lower()
                        movie_review[name] = rdict
        except:
            print("File cannot be found, Enter the correct file name \n")
            break

        revfile.close()

        while True:
            choice = input("\nWhat do you want to do?   Input 1 for similarity between two reviewers, OR Input 2 for similarity between one reviewer and all others in the database OR 3 to quit: ")  # Asking user for input

            if choice == "1":  # Option one code
                rev1 = input("Provide Reviewer1 name: ")  # Getting 2 names
                rev1 = (rev1.strip()).lower()
                rev2 = input("Provide Reviewer2 name: ")
                rev2 = (rev2.strip()).lower()
                try:
                    dic1 = movie_review[rev1]         # checking for reviewer1 in dict
                except:
                    print("Unable to find reviewer1")
                    break

                try:
                    dic2 = movie_review[rev2]        # checking for reviewer2 in dict
                except:
                    print("Unable to find reviewer2")
                    break

                set_dic1 = set(dic1)         
                set_dic2 = set(dic2)

                similar_keys = set_dic1.intersection(set_dic2)
                val = 0.0
                for r in similar_keys:             # find the similar value 
                    val = pow((float(dic1[r])-float(dic2[r])),2)+val         # euclidean distance calculation
                euclidean_dist = round(sqrt(val),3)     #upto 3 decimal places

                print("The similarity score between %s and %s is:" % (rev1.title(), rev2.title()), euclidean_dist,"\n") # print answer
            elif choice == "2": # option 2
                rev = input("Provide Reviewer name: ")     # reviewer name from user
                rev = (rev.strip()).lower()                # changing to lower case
                try:
                    dictrev = movie_review[rev]
                except:
                    print("Unable to find reviewer")      # exception catcher for reviewer name
                    break
                set_dictrev = set(dictrev)
                i = 0
                print("The Similarity Scores are:")
                for r in movie_review:
                    if r != rev:
                        comparedict = movie_review[r] 
                        set_comparedict = set (comparedict)
                        similar_keys = set_dictrev.intersection(set_comparedict)
                        val = 0.0
                        for i in similar_keys:       # find the similar value with respect to others using a loop
                            val = pow((float(dictrev[i]) - float(comparedict[i])), 2) + val # euclidean distance calculation
                        euclidean_dist = round(sqrt(val),3)   #upto 3 decimal places
                        print(rev.title(),"\t",(r.title()).ljust(16),"\t",euclidean_dist)
            elif choice == "3":        # Option 3, exit code
                print("Goodbye!")
            else:
                print("Wrong option,choose the correct option")
                continue


if __name__ == "__main__":
    main()
```
