```python
import urllib.request, urllib.parse, urllib.error
from bs4 import BeautifulSoup
import ssl
import requests
import re
import nltk
from nltk.corpus import stopwords
nltk.download('stopwords')
from nltk.tokenize import word_tokenize

# Ignore SSL certificate errors
ctx = ssl.create_default_context()
ctx.check_hostname = False
ctx.verify_mode = ssl.CERT_NONE

headers = {'User-Agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_13_6) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/70.0.3538.77 Safari/537.36'}
```

```python
url = 'https://www.imdb.com/chart/top/?ref_=nv_mv_250'
page = requests.get(url, headers = headers)
soup = BeautifulSoup(page.content, "html.parser")
```

```python
container = soup.find_all('td', class_ = 'titleColumn')
print(type(container))
```

```python
len(container)
```

```python
movie_names = []
movies = container[:50] #here we get the top 50 movies we want
for movie in movies:
    name = movie.find('a').text
    movie_names.append(name)
print(movie_names)
```

```python
movie_links = []
movies = container[:50]
for movie in movies:
    tag = movie.find('a')
    link = tag.get('href', None)
    movie_links.append(link)
print(movie_links)
```

```python
for i in range(len(movie_links)):
    movie_links[i] = 'https://www.imdb.com'+ movie_links[i]
print(movie_links)
```

```python
genre_lst = []
for i in movie_links:
    m_page_meta = requests.get(i, headers = headers)
    m_soup_meta = BeautifulSoup(m_page_meta.content, 'lxml')
    genre = m_soup_meta.find('div', class_ = 'subtext')
    genre_meta_p = genre.find('a').text.lower().strip()
    #genre_meta_f = re.sub('Genre\(s\):','',genre_meta_p).lower()
    genre_meta = re.sub('\s+','',genre_meta_p).split(',')
    genre_lst.append(genre_meta)
    
print(movie_links)
print(genre_lst)
```
